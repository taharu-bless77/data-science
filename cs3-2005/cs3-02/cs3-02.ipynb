```python
import numpy as np
import pandas as pd
```

```python
csv_in = 'wholesale_customers.csv'
df = pd.read_csv(csv_in, sep=',',skiprows=13,header=0)
print(df.shape)
print(df.info())
display(df.head())
```

```python
ser_region = df['Region']
print(type(ser_region))
print(ser_region)
```

```python
df2 = df[['Region','Milk','Grocery']]
print(type(df2))
display(df2.head())
```

```python
df3 = df.loc[1:3, 'Milk':'Frozen']
print(type(df3))
display(df3)
```

```python
df4 = df.loc[:,'Milk':]
#行方向は先頭から末尾まで
#列方向はMilk列から末尾まで
display(df4)
```

```python
df5 = df.loc[:,:'Frozen']
display(df5)
```

```python
df6 = df.iloc[1:4,3:6]
#行番号1から4の1つ手前（行番号3まで）
#列番号3から6の1つ手前（列番号5まで）
display(df6)
```
